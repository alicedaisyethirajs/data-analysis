```python
#duplicates
df.duplicated(subset=['patientid','no_show']).sum()
```

```python
#changing the dtypes for analysis
df["patientid"] = df["patientid"].astype(str)
df["appointmentid"] = df["appointmentid"].astype(str)
```

```python
# we see that the schedule daya and appointment day are in onject format we gotto change that to datetime
df["scheduledday"]=pd.to_datetime(df["scheduledday"])
df["appointmentday"]=pd.to_datetime(df["appointmentday"])
```

```python
df
```

```python
# since we are dealing with appointments lets check the lead time , which means the duration between the appointment and when it takes place
# since we are dealing with appointments lets check the lead time , which means the duration between the appointment and when it takes place
df["lead_time"]=(pd.to_datetime(df['appointmentday'])-pd.to_datetime(df['scheduledday'])).dt.days
```

```python
df["lead_time"]
```

```python
# we see that there are negative time, which is impossible so lets clean it
df[df["lead_time"]<0]
```

```python
df=df[df["lead_time"]>=0].reset_index(drop=True)
```

```python
# lets visualize our target
df["no_show"].value_counts().plot(kind="barh",figsize=(8,6))
plt.xlabel("Count")
plt.ylabel("Target Variable")
plt.title("No Show Count(target) per category")
plt.show()
```

```python
# since we have age lets check the min and max age in the record
print("min age",df["age"].min())
print("max age",df["age"].max())
```

```python
# bins is a single range of continuous values used to group values in a chart.
# lets group the tenure in bins of 12 months
labels=["{0}-{1}".format(i,i+20) for i in range (1,118,20)]
df['age_group']=pd.cut(df.age,range(1,130,20),right=False,labels=labels)
```

```python
df
```

```python
# since we have the age group we can drop age
df.drop("age",axis=1,inplace=True)
```

```python
# lets convert our target variables into 0 and 1
df["no_show"]=np.where(df["no_show"]=="No",0,1)
```

```python
# now lets do one hot encoding using get dummies for better analysis
df_dumies=pd.get_dummies(df)
df_dumies.head()
```

```python
# lets visualise the correlation bw the columns and the target
plt.figure(figsize=(20,10))
df_dumies.corr()["no_show"].sort_values(ascending=False).plot(kind="bar")
```
